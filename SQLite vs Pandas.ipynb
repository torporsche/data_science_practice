```python
#Top Make
c = con.cursor()
sql = "SELECT Make, count(Make) as Count FROM uncorrupted GROUP BY Make Order By Count Desc limit 25"
c.execute(sql)
%time rows = c.fetchall()
rows
```

```python
### Most common color by Make
sql = "SELECT Make, Color, count(Color) as Count FROM uncorrupted GROUP BY Make Order By Make, Color Desc"
c.execute(sql)
%time rows = c.fetchall()
rows
```

#  Determine if Out of State More Likely to Have Out of State Plates

```python
### Calculate residency
p = lambda x: 'In State' if  x == 'CA' else 'Out of State'

tickets['Residency'] = tickets['RP State Plate'].apply(p)


### Transfrom dates
tickets['Issue Date Fixed'] = pd.to_datetime(tickets['Issue Date'], infer_datetime_format = True)
tickets['Plate Expiry Date Num'] = pd.to_numeric(tickets['Plate Expiry Date']).fillna(0).astype(np.int)

### In converting Plate Expiry to a date, I discoverd there many invalid dates
### Drop Plate Expiry Dates before 1990 (judgement call) and after 2019 and records with month > 12 or =0

tickets = tickets[(tickets['Plate Expiry Date Num'] < 202001) & (tickets['Plate Expiry Date Num'] > 198912) & (tickets['Plate Expiry Date Num'] % 100 < 13) & (tickets['Plate Expiry Date Num'] % 100 > 0)]

### Now that valid dates are remaing, complete converting Plate Expiry to datetime
tickets['Plate Expiry Date Fixed'] = pd.to_datetime(tickets['Plate Expiry Date Num'], format='%Y%m')

### Create expired plate flag 
tickets['Expired Plates'] = np.where(tickets['Issue Date Fixed'] >= tickets['Plate Expiry Date Fixed'],'Yes','No')
```

```python
res = len(tickets['Residency'][tickets['Residency']=='In State'])
nonres = len(tickets['Residency'][tickets['Residency']=='Out of State'])

res_exp = len(tickets['Residency'][(tickets['Residency']=='In State') & (tickets['Expired Plates']=='Yes')])
nonres_exp = len(tickets['Residency'][(tickets['Residency']=='Out of State') & (tickets['Expired Plates']=='Yes')])
```
